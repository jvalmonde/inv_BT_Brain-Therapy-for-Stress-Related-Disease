```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import joblib

from matplotlib.pyplot import cm
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as lda
```

```python
features = joblib.load("/mnt/shared_drive/data/brain_therapy/features.pkl")
```

```python
features.head()
```

```python
participants = features.index.get_level_values('Participant').unique()
len(participants)
```

```python
features_df = features.loc(axis = 0)[:,:,['2D','3D']]
features_df.replace([np.inf, -np.inf], np.nan, inplace = True)
columns = features_df.columns
for column in columns:
    features_df[column]=features_df[column].ffill()
#     features_df[column]=features_df[column].fillna(0)
    
# columns
# features_df = features_df.ffill()
```

```python
f, axes = plt.subplots(9,2, figsize=(17,40), dpi=100)
ax = axes.flatten()
for i, participant in enumerate(participants):
    data = features_df.loc(axis = 0)[participant,:,:]

    l = data.index.get_level_values("Type").values
    labels = data.index.get_level_values("Type").unique()
    y = pd.get_dummies(l).values.argmax(1)
    
    
    x = TSNE(n_components=2).fit_transform(data.values)

    ax[i].set_title("Participant-{}".format(participant), size = 17)
    ax[i].plot(x[:,0][y==0],x[:,1][y==0],'o',mec='black',mew=0.5,label =labels[0])
    ax[i].plot(x[:,0][y==1],x[:,1][y==1],'o',mec='black',mew=0.5, label = labels[1])
#     break
    ax[i].legend()
plt.tight_layout()
plt.show()
```

```python
l = features_df.index.get_level_values("Type").values
labels = features_df.index.get_level_values("Type").unique()
y = pd.get_dummies(l).values.argmax(1)


x = TSNE(n_components=2).fit_transform(features_df.values)
plt.figure(figsize = (8,8),dpi = 100)
plt.title("All Participants", size = 17)
plt.plot(x[:,0][y==0],x[:,1][y==0],'o',mec='black',mew=0.5,label =labels[0])
plt.plot(x[:,0][y==1],x[:,1][y==1],'o',mec='black',mew=0.5, label = labels[1])
plt.legend()
```

```python
plt.figure(figsize = (8,8), dpi = 100)
colors = iter(cm.tab20(np.linspace(0,1,len(participants)*2)))
for i, participant in enumerate(participants):
    data = features_df.loc(axis = 0)[participant,:,:]

    l = data.index.get_level_values("Type").values
    labels = data.index.get_level_values("Type").unique()
    y = pd.get_dummies(l).values.argmax(1)
    
    
    x = TSNE(n_components=2).fit_transform(data.values)

    
    plt.plot(x[:,0][y==0],x[:,1][y==0],'o',mec='black',mew=0.5, color = next(colors) )
    plt.plot(x[:,0][y==1],x[:,1][y==1],'o',mec='black',mew=0.5, color = next(colors) ,label ="Participnat {}".format(participant))
#     break
plt.legend()
plt.tight_layout()
plt.title("All Participants".format(participant), size = 17)
plt.show()
```

```python
fig, axes = plt.subplots(9,2, figsize=(17,40), dpi=100)
ax2 = axes.flatten()
for i, participant in zip(range(18), participants):
#     ax2[i].set_title("Hi")
# # for i,participant in enumerate(participants):
    data = features_df.loc(axis = 0)[participant,:,:]
    X = data.values
    l = data.index.get_level_values("Type").values
    labels = sorted(data.index.get_level_values("Type").unique())
#     display(sorted(labels))
    y = pd.get_dummies(l).values.argmax(1)
    clf = lda(n_components = 1)
    clf.fit(X,y)
    x = clf.transform(X)
#     #display(ax2[i])
    ax2[i].set_title(participant)
    sns.distplot(x[:,0][y==0], ax= ax2[i], color="blue", label = labels[0])
    sns.distplot(x[:,0][y==1], ax= ax2[i], color="green", label = labels[1])
    ax2[i].legend()
    
```
