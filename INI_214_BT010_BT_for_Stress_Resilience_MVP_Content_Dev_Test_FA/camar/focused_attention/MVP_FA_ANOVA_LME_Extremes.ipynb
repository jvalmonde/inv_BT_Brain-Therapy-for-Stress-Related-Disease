```r
BT_FA = read.csv("/mnt/bkt_prd_dsv_brain_therapy_raw/Features/Focused_Attention//features_for_anova_extremes3.csv")
```

```r
remove_cols<-c("X","Participant","Type","type_order","context","context_order","Type_index","Difficulty","Perceptibility","context_index")
column_names = colnames(BT_FA)
column_names <- column_names[! column_names %in% remove_cols]
```

```r
BT_FA
```

### ANOVA : NO INTERACTIONS BETWEEN FACTORS

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, " ~ Type_index + context"))
    result <- aov(formula, data = BT_FA)
        
    writeLines("\n\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    print(summary(result))
    writeLines("----------------------------------------------------------------------------------")
    
}
```

## ANOVA WITH INTERACTIONS

Full Interactions

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, " ~ Type_index + context + (Type_index * context )"))
    result <- aov(formula, data = BT_FA)
        
    writeLines("\n\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    print(summary(result))
    writeLines("----------------------------------------------------------------------------------")
    
}
```

Type and Difficulty

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, " ~ Type_index + Difficulty + Perceptibility + (Type * Difficulty)"))
    result <- aov(formula, data = BT_FA)
        
    writeLines("\n\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    print(summary(result))
    writeLines("----------------------------------------------------------------------------------")
    
}
```

Type and Perceptibility

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, " ~ Type_index + Difficulty + Perceptibility + (Type * Perceptibility)"))
    result <- aov(formula, data = BT_FA)
        
    writeLines("\n\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    print(summary(result))
    writeLines("----------------------------------------------------------------------------------")
    
}
```

Difficulty and Perceptibility

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, " ~ Type_index + Difficulty + Perceptibility + (Difficulty * Perceptibility)"))
    result <- aov(formula, data = BT_FA)
        
    writeLines("\n\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    print(summary(result))
    writeLines("----------------------------------------------------------------------------------")
    
}
```
