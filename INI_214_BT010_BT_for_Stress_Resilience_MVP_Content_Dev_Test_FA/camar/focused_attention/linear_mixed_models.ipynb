```python
import numpy as np
import pandas as pd
import joblib
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("/mnt/bkt_prd_dsv_brain_therapy_raw/Features/Focused_Attention/features_for_anova.csv")
```

```python
data.head()
```

```python
features = [x for x in data.columns if x not in ["Participant","Type","type_order","context","context_order","Type_index","Difficulty","Perceptibility"]]
```

#### Dropping Participant 20 and 1

```python
data2 = data[(data["Participant"]!=1) & (data["Participant"]!=20)]
```

```python
data2.head()
```

```python
data2.to_csv( "/mnt/bkt_prd_dsv_brain_therapy_raw/Features/Focused_Attention/features_for_anova_dropped.csv")
```

#### Linear Mixed Models

```python
for feature in features:
    md = smf.mixedlm("{} ~ Type_index + Difficulty + Perceptibility".format(feature), data2, groups = data2["Participant"])
    mdf = md.fit()
    print("\n\n\n{}".format(feature.upper()))
    print(mdf.summary())
```

## INTERACTIONS

##### Type and Difficulty interaction

```python
for feature in features:
    md = smf.mixedlm("{} ~  (Type_index + Difficulty + Perceptibility) + (Type_index * Difficulty)".format(feature), data2, groups = data2["Participant"])
    mdf = md.fit()
    print("\n\n\n{}".format(feature.upper()))
    print(mdf.summary())
```

##### Type and perceptability interaction

```python
for feature in features:
    md = smf.mixedlm("{} ~ (Type_index + Difficulty + Perceptibility) + (Type_index * Perceptibility)".format(feature), data2, groups = data2["Participant"])
    mdf = md.fit()
    print("\n\n\n{}".format(feature.upper()))
    print(mdf.summary())
```

##### Difficulty and Perceptibility interaction

```python
for feature in features:
    md = smf.mixedlm("{} ~ (Type_index + Difficulty + Perceptibility) + ( Difficulty * Perceptibility)".format(feature), data2, groups = data2["Participant"])
    mdf = md.fit()
    print("\n\n\n{}".format(feature.upper()))
    print(mdf.summary())
```
