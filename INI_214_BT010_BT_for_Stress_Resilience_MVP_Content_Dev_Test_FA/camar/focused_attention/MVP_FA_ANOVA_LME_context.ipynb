```r
BT_FA  = read.csv("/mnt/bkt_prd_dsv_brain_therapy_raw/Features/Focused_Attention//features_for_anova.csv")
```

```r
library(lme4)
library(lmerTest)
```

```r
remove_cols<-c("Participant","Type","type_order","context","context_order","Type_index","Difficulty","Perceptibility")
column_names = colnames(BT_FA)
column_names <- column_names[! column_names %in% remove_cols]
```

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, " ~ type_order + context_order "))
    result <- aov(formula, data = BT_FA)
        
    writeLines("\n\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    print(summary(result))
    writeLines("----------------------------------------------------------------------------------")
    
}
```

```r
for (column_name in column_names){
    formula <- formula(paste(column_name, "~type_order + context_order + (1 | Participant)"))
    lm_model <- lmer(formula, data = BT_FA)
    
    writeLines("\n\n")
    writeLines("----------------------------------------------------------------------------------")
    writeLines(paste("\t\t\t\t",column_name))
    writeLines("----------------------------------------------------------------------------------")
    writeLines("\n")
    print(summary(lm_model)$coefficients)
}
```

```r

formula1 <- formula(paste("skin_temperature_sem", " ~ Type_index + type_order "))
result1 <- aov(formula1, data = BT_FA)

writeLines("\n\n\n")
writeLines("----------------------------------------------------------------------------------")
writeLines(paste("\t\t\t\t","skin_temperature_sem"))
writeLines("----------------------------------------------------------------------------------")
print(summary(result1))
writeLines("----------------------------------------------------------------------------------")


formula2 <- formula(paste("skin_temperature_sem", " ~ Type_index * type_order "))
result2 <- aov(formula2, data = BT_FA)
writeLines("\n\n\n")
writeLines("----------------------------------------------------------------------------------")
writeLines(paste("\t\t\t\t","skin_temperature_sem"))
writeLines("----------------------------------------------------------------------------------")
print(summary(result2))
writeLines("----------------------------------------------------------------------------------")


result <- anova(result1,result2)

writeLines("\n\n\n")
writeLines("----------------------------------------------------------------------------------")
writeLines(paste("\t\t\t\t","skin_temperature_sem"))
writeLines("----------------------------------------------------------------------------------")
print(result)
writeLines("----------------------------------------------------------------------------------")

```
