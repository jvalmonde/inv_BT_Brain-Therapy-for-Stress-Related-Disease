# Generating raw Dataframe

```python
log_events = joblib.load("/mnt/shared_drive/data/brain_therapy/log_events.pkl")
```

```python
path = "/mnt/shared_drive/data/brain_therapy/Brain_Therapy_MVP_Content_Pilot/FA_corrected_signals"
files = os.listdir(path)
```

```python
sensor_data = pd.DataFrame([])
for file in files:
    data = pd.read_csv(os.path.join(path,file))
    data['Participant'] = file.split("-")[3].split(".")[0]
    data['Datetime'] = pd.to_datetime(data['Datetime'])
    data.set_index(['Participant','Datetime'], inplace = True)
    sensor_data = pd.concat([sensor_data,data])
```

```python
participants = sensor_data.index.get_level_values('Participant').unique().values
participants = sorted(participants)
sensor_data.sort_index(sort_remaining = True, inplace = True)
sensor_data.head()
```

```python
log_events = log_events.reset_index()
log_events.rename(columns = {"participant":"Participant","time":"Datetime"}, inplace = True)
log_events.set_index(["Participant","Datetime"], inplace = True)
```

```python
log_events.head()
```

```python
sensor_data_merged = sensor_data.merge(log_events, how = "left", left_index = True, right_index = True)
```

```python
sensor_data_merged['type'] = sensor_data_merged['type'].ffill()
sensor_data_merged['context'] = sensor_data_merged['context'].ffill()
sensor_data_merged.drop("index", axis= 1, inplace = True)
```
